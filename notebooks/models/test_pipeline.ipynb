```python
import nbformat
# Função para importar uma célula de código do notebook
def import_from_notebook(notebook_path, cell_index):
    with open(notebook_path, 'r') as f:
        notebook = nbformat.read(f, as_version=4)
    code = notebook.cells[cell_index].source
    exec(code, globals())

# Caminho para o notebook (ajuste conforme necessário)
notebook_path = r'../pipe/pipeline.ipynb'  # Usando uma string bruta

# Importa a célula de código que define o CustomTransformer
import_from_notebook(notebook_path, cell_index=0) 
import pickle
```

Importe aqui seu Dataset:

```python
import pandas as pd;
```

```python
df = pd.read_csv("fraudValidate.csv")
```

Mostre do jeito que preferir os dados que você acha mais relevante

```python
df[["merchant" ,"city","is_fraud" , "amt","trans_date_trans_time" ,"is_fraud" ]]
```

Faça agrupamento de uma coluna á sua escolha usando o Groupby

```python
df.groupby(["merchant" ,"amt","trans_num"], as_index=False)["is_fraud"].sum()
```

Faça agrupamento de mais de uma coluna á sua escolha usando o Groupby

```python
df.groupby("city", as_index=False)["is_fraud"].sum()
```

Coloque o DataFrame resultante em ordem decrescente

```python
df.groupby("city", as_index=False)["is_fraud"].sum().sort_values(by="is_fraud",ascending=False)
```
